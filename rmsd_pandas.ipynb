# RMSD Plotting under Pandas Dataframe 

```python
from pandas import DataFrame, read_csv
import matplotlib.pyplot as plt
import pandas as pd
import sys
import matplotlib
%matplotlib inline
```

## 4NPQ_BA1 WT

```python
Location_4NPQ_BA1 = '/media/scottzhuang/data1/MD/BA1/4NPQ_BA1_md.skip10.ca.rmsd.log'
rmsd_4NPQ_BA1 = pd.read_table(Location_4NPQ_BA1,sep=" ")
rmsd_4NPQ_BA1.plot(x='Frame', y='RMSD')
rmsd_4NPQ_BA1.plot.hist(x='Frame',bins=100)
```

## Ensemble

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
rmsd_4NPQ_BA1.plot(x='Frame', y='RMSD', ax=axes[0,0],legend=False,title="WT")
rmsd_4NPQ_BA1_F238L.plot(x='Frame', y='RMSD', ax=axes[0,1],legend=False,title="F238L")
rmsd_4NPQ_BA1_I233T.plot(x='Frame', y='RMSD', ax=axes[1,0],legend=False,title="I233T")
rmsd_F238LI233T.plot(x='Frame', y='RMSD', ax=axes[1,1],legend=False,title="F238LI233T")
fig.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1)
rmsd_4NPQ_BA1.plot(x='Frame', y='RMSD', ax=axes,alpha=0.7)
rmsd_4NPQ_BA1_F238L.plot(x='Frame', y='RMSD', ax=axes,alpha=0.7)
rmsd_4NPQ_BA1_I233T.plot(x='Frame', y='RMSD', ax=axes,alpha=0.7)
rmsd_F238LI233T.plot(x='Frame', y='RMSD', ax=axes,alpha=0.7)
axes.legend(['WT', 'F238L','I233T','I233TF238L'])
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
rmsd_4NPQ_BA1.plot.hist(x='Frame',bins=100,ax=axes[0,0],legend=False,title="WT")
rmsd_4NPQ_BA1_F238L.plot.hist(x='Frame',bins=100,ax=axes[0,1],legend=False,title="F238L")
rmsd_4NPQ_BA1_I233T.plot.hist(x='Frame',bins=100,ax=axes[1,0],legend=False,title="I233T")
rmsd_F238LI233T.plot.hist(x='Frame',bins=100,ax=axes[1,1],legend=False,title="F238LI233T")
fig.tight_layout()
```
