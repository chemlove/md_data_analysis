```python
def create_Hydrogen_bond_analysis_data(traj_note, location = '/media/scottzhuang/data/MD/', skip=10):
    top_location = traj_note + '/' + traj_note + ".system.pdb"
    traj_location = traj_note + '/' + traj_note + ".skip" + str(skip) + ".system.xtc"
    traj = md.load(location + traj_location,top = location + top_location)
    topology = traj.topology
    table, bonds = topology.to_dataframe()
    n_frames = traj.n_frames
    hbond_analysis_dataframe = pd.DataFrame(columns=list(['traj_time',
                                                          'A1A1','A1A2','A1A3','A1A4','A2A2','A2A3','A2A4','A3A3','A3A4','A4A4',
                                                          'A1B1','A1B2','A1B3','A1B4','A2B1','A2B2','A2B3','A2B4','A3B1','A3B2','A3B3','A3B4','A4B1','A4B2','A4B3','A4B4']))
    for time in range(0,n_frames*10000,10000):  
        interaction = {'traj_time' : time,
                          'A1A1' : 0,'A1A2' : 0,'A1A3' : 0,'A1A4' : 0,
                           'A2A2': 0,'A2A3' : 0,'A2A4' : 0,'A3A3' : 0,'A3A4' : 0,'A4A4' : 0,
                            'A1B1' : 0,'A1B2' : 0,'A1B3' : 0,'A1B4' : 0,
                           'A2B1' : 0,'A2B2' : 0,'A2B3' : 0,'A2B4' : 0,
                           'A3B1' : 0,'A3B2' : 0,'A3B3' : 0,'A3B4' : 0,
                           'A4B1' : 0,'A4B2' : 0,'A4B3' : 0,'A4B4' : 0}
        for resid in range(198,315):
            hbond_dataframe = pd.read_csv(location + traj_note + '/' + str(resid) + '_hbond_list.csv')
            hbond_dataframe['counterpart'] = hbond_dataframe['counterpart'].apply(lambda x: ast.literal_eval(x))
            water_bridge_dataframe = pd.read_csv(location + traj_note + '/' + str(resid) + '_water_bridge_list.csv')
            water_bridge_dataframe['counterpart'] = water_bridge_dataframe['counterpart'].apply(lambda x: ast.literal_eval(x))

        
            
            # Analyze percentage
            
            for residue_id in set(hbond_dataframe['resid']):

                counterpart_residue = hbond_dataframe[(hbond_dataframe['resid'] == residue_id) & (hbond_dataframe['traj_time'] == time)]['counterpart']
                for counterpart_residue_list in counterpart_residue:
                    for counterpart_residue_id in counterpart_residue_list:
                        if table.loc[counterpart_residue_id].resSeq <= 219 & table.loc[counterpart_residue_id].resSeq >= 198 :
                            helix1 = 1
                        elif table.loc[counterpart_residue_id].resSeq <= 243 & table.loc[counterpart_residue_id].resSeq >= 220:
                            helix1 = 2
                        elif table.loc[counterpart_residue_id].resSeq <= 282 & table.loc[counterpart_residue_id].resSeq >= 254:
                            helix1 = 3
                        elif table.loc[counterpart_residue_id].resSeq >= 286:
                            helix1 = 4
                        else:
                            helix1 = 0
                        if table.loc[residue_id].resSeq <= 219 & table.loc[residue_id].resSeq >= 198 :
                            helix2 = 1
                        elif table.loc[residue_id].resSeq <= 243 & table.loc[residue_id].resSeq >= 220:
                            helix2 = 2
                        elif table.loc[residue_id].resSeq <= 282 & table.loc[residue_id].resSeq >= 254:
                            helix2 = 3
                        elif table.loc[residue_id].resSeq >= 286:
                            helix2 = 4  
                        else:
                            helix2 = 0
                        if helix1 * helix2 != 0:
                            if table.loc[counterpart_residue_id].chainID == table.loc[residue_id].chainID:
                                interaction['A' + str(min(helix1,helix2)) + 'A' + str(max(helix1,helix2))] = interaction['A' + str(min(helix1,helix2)) + 'A' + str(max(helix1,helix2))] + 1
                            else:
                                interaction['A' + str(helix1) + 'B' + str(helix2)] = interaction['A' + str(helix1) + 'B' + str(helix2)] + 1
            for residue_id in set(water_bridge_dataframe['resid']):
                counterpart_residue = water_bridge_dataframe[(water_bridge_dataframe['resid'] == residue_id) & (water_bridge_dataframe['traj_time'] == time)]['counterpart'].values
                for counterpart_residue_list in counterpart_residue:
                    for counterpart_residue_id in counterpart_residue_list:
                        if table.loc[counterpart_residue_id].resSeq <= 219 & table.loc[counterpart_residue_id].resSeq >= 198 :
                            helix1 = 1
                        elif table.loc[counterpart_residue_id].resSeq <= 243 & table.loc[counterpart_residue_id].resSeq >= 220:
                            helix1 = 2
                        elif table.loc[counterpart_residue_id].resSeq <= 282 & table.loc[counterpart_residue_id].resSeq >= 254:
                            helix1 = 3
                        elif table.loc[counterpart_residue_id].resSeq >= 286:
                            helix1 = 4
                        else:
                            helix1 = 0
                        if table.loc[residue_id].resSeq <= 219 & table.loc[residue_id].resSeq >= 198 :
                            helix2 = 1
                        elif table.loc[residue_id].resSeq <= 243 & table.loc[residue_id].resSeq >= 220:
                            helix2 = 2
                        elif table.loc[residue_id].resSeq <= 282 & table.loc[residue_id].resSeq >= 254:
                            helix2 = 3
                        elif table.loc[residue_id].resSeq >= 286:
                            helix2 = 4
                        else:
                            helix2 = 0
                        if helix1 * helix2 != 0:
                            if table.loc[counterpart_residue_id].chainID == table.loc[residue_id].chainID:
                                interaction['A' + str(min(helix1,helix2)) + 'A' + str(max(helix1,helix2))] = interaction['A' + str(min(helix1,helix2)) + 'A' + str(max(helix1,helix2))] + 1
                            else:
                                interaction['A' + str(helix1) + 'B' + str(helix2)] = interaction['A' + str(helix1) + 'B' + str(helix2)] + 1    
                        
        hbond_analysis_dataframe = hbond_analysis_dataframe.append(interaction,ignore_index=True)
    hbond_analysis_dataframe.to_csv(location + traj_note + '/hbond_analysis_helix.csv')
    print('finishing!')
   #         print('At time ' + str(time) + ', in total ' + str(n_hbond) + ' hydrogen bonds between residues, resid '
   #               + str(resid) + ' forms ' + str(n_intra_hbond) + ' intra-subunit H-bond and ' + str(n_inter_hbond) + ' inter-subunit H-bond.')
```
